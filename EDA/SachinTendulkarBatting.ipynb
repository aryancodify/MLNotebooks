```python
import pandas as pd

sachin_df = pd.read_csv('D:/upgrad/EDA/tendulkar_ODI.csv')
```

```python
sachin_df.head()
```

```python
sachin_df.info()
```

```python
sachin_df[sachin_df['4s']=='-']
```

```python
sachin_df['4s'].replace("-","0",inplace=True)
```

```python
sachin_df['4s'] = sachin_df['4s'].apply(lambda x: int(x))
```

```python
sachin_df['Runs'].replace("DNB","0",inplace=True)
```

```python
sachin_df['Runs'].replace("TDNB","0",inplace=True)
```

```python
sachin_df.Runs.replace({'\*':''}, regex=True, inplace=True)
```

```python
sachin_df['Runs'] = sachin_df['Runs'].apply(lambda x: int(x))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sachin_df['Runs'].describe()
```

```python
cuts = np.arange(0, 101, 10)
```

```python
df = pd.cut(sachin_df['Runs'],cuts).value_counts().reset_index()
```

```python
df = df.rename(columns={'index':'Runs','Runs':'frequency'})
```

```python
df
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Runs',y='frequency',data=df)
```
